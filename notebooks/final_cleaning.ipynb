```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
import math
```

```python
df = pd.read_csv (r'/Users/juanmarin/IronHack/pipeline-proj/data/comb.csv')
```

```python
df.head(5)
```

```python
#Lets start dropping some rows
#We will drop rows of players who have not played enough games or minutes
#For their PER to be considered as consistent
average_minutes = df["MP"].mean()
average_games = df["G"].mean()
average_mins_per_game = average_minutes/average_games
cut_off_value = average_mins_per_game * 15
df = df[df["MP"] > cut_off_value]
df.reset_index(drop=True, inplace=True)
print(cut_off_value)
```

```python
df.head(5)
```

```python
df.shape
```

```python
df['salary'] = df['salary'].apply(lambda x: x[1:])
df['salary'] = df['salary'].apply(lambda x: x.replace(",",""))
```

```python
df.head(5)
```

```python
#Get rid of dollar sign from salary column
df['salary'] = df['salary'].apply(lambda x: int(x))
```

```python
df.shape
```

```python
df.head(5)
```

```python
#Create new column which will equal performance per dollar spent
df["return"] = np.log10(df["PER"]*1e8 / df["salary"])
```

```python
df.head(5)
```

```python
df.sort_values(by='return', ascending=False)
```

```python
df.to_csv(r'/Users/juanmarin/IronHack/pipeline-proj/data/ordered.csv', index = False)
```
